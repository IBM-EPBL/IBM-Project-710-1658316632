# Opencv for video processing

```python
msg_sent = False
model = load_model('/home/veeraraghavan/Forest_fire.h5')


v = 0
vid= cv2.VideoCapture(r'/home/veeraraghavan/demo.mp4')
if vid.isOpened():
  while True:
    success,frame = vid.read()
    if success:
         
         cv2.imwrite('output.jpg',frame)
         s='/home/veeraraghavan/output.jpg'
         img = image.load_img(s,target_size=(150,150))
         x = image.img_to_array(img)
         x = np.expand_dims(x,axis=0)
         predict = model.predict(x)
         y = int(predict[0][0])
    
         if y==0:
              if not msg_sent:
                   send_message()
                   msg_sent=True
         else:
            print("No Forest Fire Detected")
    
    else:
       break
    cv2.imshow('frame',frame)
    print("frame")
    print(v)
    v+=1
 
    
       
  vid.release()
cv2.destroyAllWindows()
```
